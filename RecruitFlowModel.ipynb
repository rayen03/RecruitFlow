```python
from transformers import TFAutoModel , AutoTokenizer
import pandas as pd
from datasets import Dataset , DatasetDict
```

```python
# Load your CSV file into a Pandas DataFrame
df = pd.read_csv("resume_job.csv")

# Assuming your CSV has columns 'CV Text', 'Job Description', and 'Label'
columns = ['CV Text', 'Job Description', 'Label']

# Create a Hugging Face Dataset
hf_dataset = Dataset.from_pandas(df[columns])

# Split the dataset into train, test, and validation
splits = hf_dataset.train_test_split(test_size=0.2)
dataset_dict = DatasetDict({
    'train': splits['train'],
    'test': splits['test'],
    'validation': splits['test']  # You can adjust this based on your needs
})

## this bloc will turn the csv file into a the hugging's face dataset dict format to work with it
```

```python
model = TFAutoModel.from_pretrained("bert-base-uncased")
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
def tokenize(batch):
    return tokenizer(batch["CV Text"], padding=True, truncation=True)
```

```python
dataset_encoded = dataset_dict.map(tokenize, batched=True, batch_size=None)
```

```python
import tensorflow as tf
```

```python
dataset_encoded
```

```python
# setting 'input_ids', 'attention_mask', 'token_type_ids', and 'label'
# to the tensorflow format. Now if you access this dataset you will get these
# columns in `tf.Tensor` format

dataset_encoded.set_format('tf',
                            columns=['input_ids', 'attention_mask', 'token_type_ids', 'Label'])

# setting BATCH_SIZE to 64.
BATCH_SIZE = 12

def order(inp):
    '''
    This function will group all the inputs of BERT
    into a single dictionary and then output it with
    labels.
    '''
    data = list(inp.values())
    return {
        'input_ids': data[1],
        'attention_mask': data[2],
        'token_type_ids': data[3]
    }, data[0]


# converting train split of `emotions_encoded` to tensorflow format
train_dataset = tf.data.Dataset.from_tensor_slices(dataset_encoded['train'][:])
# set batch_size and shuffle
train_dataset = train_dataset.batch(BATCH_SIZE).shuffle(1000)
# map the `order` function
train_dataset = train_dataset.map(order, num_parallel_calls=tf.data.AUTOTUNE)

# ... doing the same for test set ...
test_dataset = tf.data.Dataset.from_tensor_slices(dataset_encoded['test'][:])
test_dataset = test_dataset.batch(BATCH_SIZE)
test_dataset = test_dataset.map(order, num_parallel_calls=tf.data.AUTOTUNE)
```

```python
inp, out = next(iter(train_dataset)) # a batch from train_dataset
print(inp, '\n\n', out)
```

```python
class BERTForClassification(tf.keras.Model):

    def __init__(self, bert_model, num_classes):
        super().__init__()
        self.bert = bert_model
        self.fc = tf.keras.layers.Dense(num_classes, activation='softmax')

    def call(self, inputs):
        x = self.bert(inputs)[1]
        return self.fc(x)
```

```python
classifier = BERTForClassification(model, num_classes=2)

classifier.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=['accuracy']
)
```

```python
history = classifier.fit(
    train_dataset,
    epochs=3
)
```

```python
classifier.evaluate(test_dataset)
```
